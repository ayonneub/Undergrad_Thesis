```python
from typing import Tuple
import itertools 
from pandas import read_excel
suffix_list={}
prefix_list={}
counting={}
check_root={}
check_root_pref={}
counting_pref={}
class TrieNode(object):    
    def __init__(self, char: str):
        self.char = char
        self.children = []
        # Is it the last character of the word.`
        self.word_finished = False
        # How many times this character appeared in the addition process
        self.counter = 1
def add(root, word: str):
    """
    Adding a word in the trie structure
    """
    node = root
    for char in word:
        found_in_child = False
        # Search for the character in the children of the present `node`
        for child in node.children:
            if child.char == char:
                # We found it, increase the counter by 1 to keep track that another
                # word has it as well
                child.counter += 1
                # And point the node to the child that contains this char
                node = child
                found_in_child = True
                break
        # We did not find it so add a new chlid
        if not found_in_child:
            new_node = TrieNode(char)
            node.children.append(new_node)
            # And then point node to the new child
            node = new_node
    # Everything finished. Mark it as the end of a word.
    node.word_finished = True
    if f==0:
        check_root[word]=1
    else:
        check_root_pref[word[::-1]]=1
def find_prefix(root, prefix: str) -> Tuple[bool, int]:
    """
    Check and return 
      1. If the prefix exsists in any of the words we added so far
      2. If yes then how may words actually have the prefix
    """
    node = root
    # If the root node has no children, then return False.
    # Because it means we are trying to search in an empty trie
    if not root.children:
        return
    for index,char in enumerate(prefix):
        if f==0:
            if prefix[:index+1] not in counting:
                counting[prefix[:index+1]]=1
            else:
                counting[prefix[:index+1]]+=1
        else:
            temping=prefix[:index+1]
            temping=temping[::-1]
            if temping not in counting_pref:
                
                counting_pref[temping]=1
            else:
                counting_pref[temping]+=1
        
            

        
        char_not_found = True
        # Search through all the children of the present `node`
        for child in node.children:
            if child.char == char:
                # We found the char existing in the child.
                char_not_found = False
                # Assign node as the child containing the char and break
                node = child
                
                break
        # Return False anyway when we did not find a char.
        if char_not_found:
            return 
    # Well, we are here means we have found the prefix. Return true to indicate that
    # And also the counter of the last node. This indicates how many words have this
    # prefix
    return




if __name__ == "__main__":
    root = TrieNode('*')
    
    #f=open("final_data.txt", "r",encoding="utf-8")
    #f1 =f.readlines()
    test_word=[]
    #for x in f1:
        #x = x.replace('\n','')
        #x = x.replace('\t','')
        #x=x.replace(' ','')
        #test_word.append(x)
        
    
    f=0
    # find your sheet name at the bottom left of your excel file and assign 
    # it to sheet_name
    my_sheet = 'Sheet1'
    file_name = 'Bangla_Unique_Word.xlsx' # name of your excel file
    df = read_excel(file_name, sheet_name = my_sheet)
    temp=list(df.head())
    list_word=temp
    for x in temp:
        list_word=list_word+list(df[x]) 
    list_word = [word for word in list_word if str(word) != 'nan'] 
    list_word = list( dict.fromkeys(list_word) )
    
    for x in list_word:
        add(root,x)
        #print(x)
    for  x in list_word:
        find_prefix(root,x)
    
```

```python
#segmenting the word into prefix ,root and suffix

segment_pref={}
segment_root={}
segment_suff={}
#suffix detection
#test_word=['ফৌজদারি']
for x in list_word:
    i=len(x)-1
    maxi=1000000000000
    str=""
    while i>=0:
        str+=x[i]
        temp=str[::-1]
        if x[:i]=='':
            i=i-1
            continue
        prob=counting[x]/counting[x[:i]]
        #prob2=counting[x[:i]]/counting[x[:i-1]]
       # print(prob,end=' ')
        #print(x,end=' ')
        #print(x[:i])
        if  temp in suffix_list and suffix_list[temp]>20 and prob<maxi and prob<1:
                
            segment_suff[x]=temp
            maxi=prob
        i=i-1
```

```python
#finding roots
#test_word=['ছাগলটি']
for x in list_word:
    stri=''
    root=x
    for index,i in enumerate(x):
        stri+=i
        if x in segment_pref and segment_pref[x]==stri:
            
            root=x[index+1:]
    stri=''
    i=len(root)-1
    while i>=0:
        stri+=root[i]
        temp=stri[::-1]
        if x in segment_suff and segment_suff[x]==temp:
            root=root[:i]
            segment_root[x]=root
        i-=1
    
```
